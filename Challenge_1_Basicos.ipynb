```python
import pandas as pd

# Assuming the uploaded file is 'globalterrorismdb (1).csv'
df = pd.read_csv('globalterrorismdb (1).csv', encoding='ISO-8859-1')
```

```python
attacks_by_country = df.groupby("country_txt").size().sort_values(ascending=False).head(10)
```

```python
attacks_by_country.plot(kind="bar", figsize=(10,5))
plt.title("Top 10 countries with most attacks")
plt.show()
```
